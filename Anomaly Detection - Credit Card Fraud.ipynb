Since the data is very large we will just take a small sample and then work on it for the model

```python
data_sample = data.sample(frac = 0.1, random_state=1)

data_sample.shape
```

```python
data.shape
```

From the above shape we see that we have taken exactly 1% of the total data in the sample

```python
#Determining the fraud and normal transactions in the sample data

Fraud = data_sample[data_sample['Class']==1]

Normal = data_sample[data_sample['Class']==0]

outlier_fraction = len(Fraud)/float(len(Normal))
```

```python
print(outlier_fraction)
print("Fraud Cases: ", len(Fraud))
print("Normal Cases: ",len(Normal))
```

Lets check the correlation between the features

```python
columns = data_sample.columns.tolist()

#Variables that we want to include as independent variables
columnsToInclude = [c for c in columns if c not in ["Class"]]
#Variable that we are predicting
target = "Class"
#Defining a random state
state = np.random.RandomState(RANDOM_SEED)
X = data_sample[columnsToInclude]
Y = data_sample[target]

X_outliers = state.uniform(low=0, high=1, size=(X.shape[0], X.shape[0]))

print(X.shape)
print(Y.shape)
```

# Model Prediction

Isolation Forest Algorithm

Local Outlier Factor Algorithm

Support Vector Machine

```python
#Creating a dictionary for all the classifiers
classifiers = {"Isolation Forest":IsolationForest(n_estimators=100, max_samples=len(X),
                                                  contamination=outlier_fraction,random_state=state, verbose=0),
    "Local Outlier Factor":LocalOutlierFactor(n_neighbors=20, algorithm='auto', 
                                              leaf_size=30, metric='minkowski',
                                              p=2, metric_params=None, contamination=outlier_fraction),
    "Support Vector Machine":OneClassSVM(kernel='rbf', degree=3, gamma=0.1,nu=0.05, 
                                         max_iter=-1, random_state=state)
   
}
```

```python
type(classifiers)
```

```python
#running all the classifiers
n_outliers = len(Fraud)
for i, (clf_name, clf) in enumerate(classifiers.items()):
    #Fit the data and tag outliers
    if clf_name == "Local Outlier Factor":
        y_pred = clf.fit_predict(X)
        scores = clf.negative_outlier_factor_
    elif clf_name == "Support Vector Machine":
        clf.fit(X)
        y_pred = clf.predict(X)
    else:    
        clf.fit(X)
        scores_prediction = clf.decision_function(X)
        y_pred = clf.predict(X)
    #Reshape the prediction values to 0 for Valid transactions , 1 for Fraud transactions
    y_pred[y_pred == 1] = 0
    y_pred[y_pred == -1] = 1
    n_errors = (y_pred != Y).sum()
    # Run Classification Metrics
    print("{}: {}".format(clf_name,n_errors))
    print("Accuracy Score :")
    print(accuracy_score(Y,y_pred))
    print("Classification Report :")
    print(classification_report(Y,y_pred))
```
